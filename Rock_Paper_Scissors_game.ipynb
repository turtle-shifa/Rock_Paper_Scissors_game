```python
import random
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

server_list = [rock,paper,scissors]
round = int(input("Please tell us how many rounds you want to play. "))
computer = 0
human = 0
for i in range(round):
  print("What do you choose? Type 0 for Rock, 1 for Paper, or 2 for Scissors")
  user = int(input("Your turn: "))
  print(server_list[user])
  print("Computer chose")
  random_number = random.randint(0,2)
  print(server_list[random_number])
  if random_number == user:
    pass
  else:
    if user == 0:
      if random_number == 1:
        computer+=1
      else:
        human+=1
    elif user == 1:
      if random_number == 0:
        human+=1
      else:
        computer+=1
    elif user == 2:
      if random_number == 0:
        computer+=1
      else:
        human+=1

print(f"Your Score: {human}\nComputer Score: {computer}")
if human == computer:
  print(f"Result: DRAW")
elif human>computer:
  print(f"Result: YOU WON üéñÔ∏è")
else:
  print(f"Result: YOU LOSE üò•")
```
